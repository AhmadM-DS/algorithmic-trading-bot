```python
#Standard library
import sys
sys.path.append("..")
import os

#Third party libaries
import pandas as pd

#Local imports
from src.data.cleaner import fetch_raw_data, clean_data
from src.strategies.rsi import generate_signals
from src.visuals import plot_portfolio_growth
from src.backtest import run_backtest, display_results
```

```python
#Fetch and clean data for Apple, Nvidia and Microsoft stock
stocks = ["AAPL", "NVDA", "MSFT"]
all_signals = {}

for stock in stocks:
    if not os.path.exists(f"../data/cleaned/{stock}_cleaned.csv"): 
        raw_data = fetch_raw_data(stock, "2020-01-01", "2024-01-01")
        cleaned_data = clean_data(stock)
    #Generate trading signals
    else:
        cleaned_data = pd.read_csv(f"../data/cleaned/{stock}_cleaned.csv")
    all_signals[stock] = generate_signals(cleaned_data)
```

```python
#Run backtest for each stock and print results
for stock in stocks:
    results = run_backtest(all_signals[stock], stock)
    display_results(results)
    plot_portfolio_growth(results, stock)
```
