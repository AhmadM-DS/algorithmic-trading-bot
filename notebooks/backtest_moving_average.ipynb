```python
#Standard library
import sys
sys.path.append("..")

#Third party libraries
import pandas as pd
import matplotlib.pyplot as plt

#Local imports
from src.data.cleaner import fetch_raw_data, clean_data
from src.strategies.moving_average import generate_signals
from src.visuals import plot_signals
from src.backtest import run_backtest
```

```python
#Fetch and clean data for Apple, Nvidia and Microsoft stock
stocks = ["AAPL", "NVDA", "MSFT"]
all_signals = {}

for stock in stocks:
    raw_data = fetch_raw_data(stock, "2020-01-01", "2024-01-01")
    cleaned_data = clean_data(stock)
    #Generate trading signals
    all_signals[stock] = generate_signals(cleaned_data)
```

```python
#Plot the Moving Average Crossover Strategy for each stock
for stock in stocks:
    plot_signals(all_signals[stock], stock)
```

```python
#Run backtest for each stock and print results
for stock in stocks:
    results = run_backtest(all_signals[stock], stock)
    print(f"Ticker: {results['ticker']}")
    print(f"Starting Capital: ${results['starting_capital']}")
    print(f"Final Capital: ${results['final_capital']:.2f}")
    print(f"Total Return: ${results['total_return']:.2f}")
    print(f"Return Percentage: {results['percent_return']:.2f}%")
    print(f"Win Rate: {results['win_rate']:.2f}%")
    print(f"Total Trades: {len(results['trades'])}")
```
